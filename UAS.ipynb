```python
import pandas as pd
```

```python
train_df = pd.read_csv("train (1).csv")
validation_df = pd.read_csv("validation.csv")
test_df  = pd.read_csv("test.csv")
```

```python
train_df.head()
validation_df.head()
test_df.head()
```

```python
train_X = train_df.drop('class', axis=1)
train_Y = train_df['class']
```

```python
train_X
```

```python
validation_X = validation_df.drop('class', axis=1)
validation_Y = validation_df['class']
```

```python
test_X = test_df.drop('class', axis=1)
test_y = test_df['class']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import precision_recall_fscore_support as score
classifier = RandomForestClassifier()
```

```python
# default settings
classifier.fit(train_X, train_Y)
prediction = classifier.predict(validation_X)
print(classification_report(validation_Y, prediction))
```

```python
# scenario
n_estimators = [10, 50, 100]
max_depths = [1, 5, 10, 100]
criteria = ['gini', 'entropy']

for n in n_estimators:
  for depth in max_depths:
    for crit in criteria:
      new_classifier = RandomForestClassifier(n_estimators = n, max_depth=depth, criterion=crit)
      new_classifier.fit(train_X, train_Y)
      prediction = classifier.predict(validation_X)
      print('max_depth {} criterion {} n_estimator {}'.format(depth, crit, n))
      print(classification_report(validation_Y, prediction))
      print('=====')
```
